```python
import pandas as pd
import numpy as np
import seaborn as sns
df1=pd.read_csv('/content/drive/MyDrive/Titanic project/gender_submission.csv') 
df2=pd.read_csv('/content/drive/MyDrive/Titanic project/train.csv') 
print(df2)
y=df2.iloc[ :,[1]]
df2=df2.iloc[ :,[0,2,4,5,6,7,8,10,11]]
df3=pd.read_csv('/content/drive/MyDrive/Titanic project/test.csv')
# print(df1)
# del df2['Ticket']
# df2
# print(y)
df2.isna().sum()
df3=df3.iloc[ :,[0,1,3,4,5,6,7,9,10]]
data3=df3
# y[0:5]
y3=df1
submission=y3.iloc[:,[0]]
# del y3['PassengerId']
# submission
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
df2.head()
data=df2
# del data['Fare']
data.info()
```

```python
data=data.fillna(method='ffill')
data3=data3.fillna(method='ffill')

le=LabelEncoder()

# FOR TRAINING DATA
data["Sex"]=le.fit_transform(df2["Sex"])
data["Cabin"]=le.fit_transform(df2["Cabin"])
data["Embarked"]=le.fit_transform(df2["Embarked"])
data["Ticket"]=le.fit_transform(df2["Ticket"])

data3["Sex"]=le.fit_transform(df3["Sex"])
data3["Cabin"]=le.fit_transform(df3["Cabin"])
data3["Embarked"]=le.fit_transform(df3["Embarked"])
data3["Ticket"]=le.fit_transform(df3["Ticket"])


# data
print(data)
print(data3)
```

```python

sc_data=data.copy()
sc_data3=data3.copy()
# print(y)
data
# print(sc_data[])
# print(sc_data)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(sc_data)
sc_data=scaler.transform(sc_data)
sc_data3=scaler.transform(sc_data3)

from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression(max_iter=500).fit(sc_data, y.values.ravel()) 

from sklearn.metrics import confusion_matrix

y_pred=regressor.predict(sc_data3)
y_actual=y3

confusion_matrix(y_actual, y_pred)
```
